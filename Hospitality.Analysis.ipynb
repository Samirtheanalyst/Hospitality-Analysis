```python
# Cancellation rate
Cancellation_rate= f"{round(Hospitality_data['booking_status'].value_counts()/Hospitality_data['booking_id'].count()*100,2)}%"
print(Cancellation_rate)
```

```python
# Trend analysis; Booking by month
Monthly_booking= Hospitality_data['booking_date'].dt.strftime('%B').value_counts().reset_index()
Monthly_booking.columns= ['Month', 'Total_booking']
Monthly_booking= Monthly_booking.sort_values('Total_booking', ascending=False)
print(Monthly_booking)
```

```python
# Revenue By Month
Monthly_revenue = Hospitality_data.groupby(Hospitality_data['check_in_date'].dt.strftime('%B'))['revenue_realized'].sum().reset_index()
Monthly_revenue.columns = ['Month', 'Total_revenue'] 
Monthly_revenue = Monthly_revenue.sort_values('Total_revenue',ascending=True)
print(Monthly_revenue)
```

```python
# Weekday Vs Weekend Bookings
Daytype_Booking = Hospitality_data.groupby('day_type')['booking_id'].count().reset_index()
Daytype_Booking.columns=['Daytype','Total_Booking']
Daytype_Booking
```

```python
# Revenue By City
City_revenue = Hospitality_data.groupby('city')['revenue_realized'].sum().reset_index()
City_revenue.columns=['City','Total_revenue']
City_revenue=City_revenue.sort_values('Total_revenue',ascending=False).reset_index(drop=True)
City_revenue
```

```python
# Revenue By Hotel
Hotel_revenue = Hospitality_data.groupby('property_name')['revenue_realized'].sum().reset_index()
Hotel_revenue.columns=['Hotel_name','Total_revenue']
Hotel_revenue=Hotel_revenue.sort_values('Total_revenue',ascending=False).reset_index(drop=True)
Hotel_revenue
```

```python
# Class Wise Booking
Room_class_bookings = Hospitality_data.groupby('room_class')['revenue_realized'].count().reset_index()
Room_class_bookings.columns=['Room Class','Total Bookings']
#sort them in descending
Room_class_bookings = Room_class_bookings.sort_values('Total Bookings',ascending=False)
print(Room_class_bookings)
```

```python
# Class Wise Revenue
Room_class_revenue = Hospitality_data.groupby('room_class')['revenue_realized'].sum().reset_index()
Room_class_revenue.columns = ['Room Class','Total Revenue']
Room_class_revenue = Room_class_revenue.sort_values('Total Revenue',ascending=False)
print(Room_class_revenue)
```
